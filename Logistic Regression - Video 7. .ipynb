```python
# Packages for anaysis
import numpy as np
import pandas as pd

# Packages for visuals
import matplotlib.pyplot as plt
import seaborn as sns;
%matplotlib inline
# Allows charts to appear in the notebook
%matplotlib inline

# Spliting the data
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("cancer.csv")
data.drop("Unnamed: 32", axis=1, inplace=True)
```

```python
sns.jointplot("radius_mean", "texture_mean", data = data)
```

```python
sns.heatmap(data.corr()) #good for coleration
```

```python
#To check if columns have null values

data.isnull().sum() 
```

```python
x = data[['radius_worst', 'texture_worst', 'perimeter_worst', 'area_worst', 'smoothness_worst']]
y = data['diagnosis']
x.head()
```

```python
y.head()
```

```python
## Spliting the data = from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression

logModel = LogisticRegression()
logModel.fit(x_train, y_train)
```

```python
y_pred = logModel.predict(x_test)
```

```python
print(y_pred)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
